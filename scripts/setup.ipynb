```python
import pandas as pd
pd.set_option('display.max_rows', None)
import numpy as np
RAW_DATA = "../raw_data/"
```

```python
df_institucional = pd.read_excel(RAW_DATA + "DadosAvInstitucional_2025.xlsx", sheet_name=None)
df_cursos = pd.read_excel(RAW_DATA + "DadosAv_Cursos_2024.xlsx", sheet_name=None)
df_presenciais = pd.read_excel(RAW_DATA + "DadosAvDisciplinasPresenciais_1S2025.xlsx", sheet_name=None)
df_ead = pd.read_excel(RAW_DATA + "DadosAvDisciplinasEAD_1S2025.xlsx", sheet_name=None)
```

Institucional


```python
df_institucional.keys()
df_institucional['Perguntas'] = df_institucional['Perguntas'].drop(df_institucional['Perguntas'].index[75:])

df_institucional['Perguntas'][['ID_PERGUNTA', 'ID_QUESTIONARIO', 'Ordem']] = df_institucional['Perguntas'][['ID_PERGUNTA', 'ID_QUESTIONARIO', 'Ordem']].astype('int')

df_institucional['PESQUISA 442']['SIGLA_LOTACAO'].fillna(value='Não disponível', inplace=True)
```

Cursos

```python
df_cursos.keys()
df_cursos['Perguntas'] = df_cursos['Perguntas'].drop(df_cursos['Perguntas'].index[33:])
df_cursos['Perguntas'][['ID_PERGUNTA', 'ID_QUESTIONARIO', 'Ordem']] = df_cursos['Perguntas'][['ID_PERGUNTA', 'ID_QUESTIONARIO', 'Ordem']].astype('int')
df_cursos['DadosAvCursos']['RESPOSTA'].fillna(value='Sem resposta', inplace=True)
```

Presenciais

```python
df_presenciais.keys()

df_presenciais['Perguntas'] = df_presenciais['Perguntas'].drop(df_presenciais['Perguntas'].index[33:])
df_presenciais['Perguntas'][['ID_PERGUNTA', 'ID_QUESTIONARIO', 'Ordem']] = df_presenciais['Perguntas'][['ID_PERGUNTA', 'ID_QUESTIONARIO', 'Ordem']].astype('int')
```

Salvar

```python
def save_to_csv(dicts):
    CLEAN_PATH = '../clean_data/'

    for keys_dict in dicts.keys():
        for key in dicts[keys_dict].keys():
            name = str(keys_dict) + "_" + str(key).lower() + '.csv'
            
            dicts[keys_dict][key].to_csv(CLEAN_PATH + name, index=False)    
```

```python
dicts = {"institucional": df_institucional,
        "cursos": df_cursos,
        "presenciais": df_presenciais,
        "ead": df_ead
        }
```

```python
save_to_csv(dicts)
```
